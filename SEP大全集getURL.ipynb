```python
import os
import requests
from bs4 import BeautifulSoup
import time
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'}
```

```python
url = 'https://plato.stanford.edu/archives/sum2021/contents.html'
r = requests.get(url,headers=headers)
```

```python
li_list = []
text = r.text
soup = BeautifulSoup(text, 'html.parser')
lis = soup.find_all(name='li')
for li in lis:
    li_dict = {}
    li_text = li.text
    if '— see' in li_text:
        continue
    a = li.find(name='a')
    try:
        k = a.text
        v = a.get('href')
        if v.startswith('entries'):
            li_dict[k] = v
            li_list.append(li_dict)
    except:
        pass
```

```python
len(li_list)
```

```python
url_dict = {}
for li in li_list:
    for k,v in li.items():
        url_dict[v] = k
```

```python
len(url_dict)
```

```python
for url,entry in url_dict.items():
    print(entry)
    print('\n\n')
```

```python
with open('sum2021urls.txt','w',encoding='utf8') as s:
    lines = []
    for url,entry in url_dict.items():
        line = entry + '\t' + url
        lines.append(line)
    s.write('\n'.join(lines))
```
